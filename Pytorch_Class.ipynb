Element-wise Functions

Matrix Operations

```python
A = torch.randn(3,3) #creates a 3x3 tensor that follows a normal distribution
A
```

```python
B = torch.randn(3,4)
B
```

```python
matmul = torch.matmul(A,B)
matmul
```

```python
v1 = torch.tensor([1.0, 2.0, 3.0])
v2 = torch.tensor([4.0, 5.0, 6.0])
```

```python
dot_result = torch.dot(v1, v2)
dot_result
```

```python
#torch.mm
A = torch.tensor([[1., 2.],
                  [3., 4.]])  # Shape: (2, 2)


B = torch.tensor([[5., 6.],
                  [7., 8.]])  # Shape: (2, 2)
```

```python
C = torch.mm(A, B)
C
```

```python
#torch.mm only works for 2D array
A = torch.tensor([[[1., 2.], [2., 3.]],
                  [[4., 6.],[3., 4.]]])  # 3D array


B = torch.tensor([[[5., 6.],[1., 2.]],
                  [[5., 7.],[7., 8.]]])  # 3D array
```

```python
A.shape
```

```python
C = torch.mm(A, B)
C
```
