```python
def cosine_similarity(vec1, vec2):
    """
    Calculate cosine similarity between two vectors.
    
    Returns a score between -1 and 1 (higher = more similar)
    """
    dot_product = np.dot(vec1, vec2)
    norm1 = np.linalg.norm(vec1)
    norm2 = np.linalg.norm(vec2)
    return dot_product / (norm1 * norm2)
```

```python
from sentence_transformers import SentenceTransformer, util

# Load model
model = SentenceTransformer("all-MiniLM-L6-v2")

sentences = [
    "The dog is playing in the park",          # Sentence 1 (index 0)
    "A puppy is running outside",
    "The cat is sleeping on the couch",
    "Python is a programming language",        # Sentence 4 (index 3)
    "Machine learning models need data",
    "I love coding in Python"
]

# Generate embeddings
embeddings = model.encode(sentences, convert_to_tensor=True)

def compute_similarity(query_index):
    query_sentence = sentences[query_index]
    query_embedding = embeddings[query_index]

    similarities = []

    for i, sentence in enumerate(sentences):
        score = util.cos_sim(query_embedding, embeddings[i]).item()
        similarities.append((sentence, score))

    # Sort by similarity score (descending)
    similarities.sort(key=lambda x: x[1], reverse=True)

    print(f"\nQuery Sentence: '{query_sentence}'")
    print("-" * 60)
    for sent, score in similarities:
        print(f"{score:.4f}  |  {sent}")

    return similarities


# Sentence 1 vs all others
sim_sentence_1 = compute_similarity(query_index=0)

# Sentence 4 vs all others
sim_sentence_4 = compute_similarity(query_index=3)
```
